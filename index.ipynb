## Compare subway traffic by day of the week. Display this as a graph.

```python
# Let's start by checking format of our 'DATE' column.
df.DATE.dtype
```

```python
# Let's take a look:
df.DATE.iloc[0]
```

```python
# We can use pandas to convert into a format we can manipulate.
df.DATE = pd.to_datetime(df.DATE)
```

```python
df.DATE.iloc[0]
```

```python
# Adding in datevalue, index begins at 0/Sunday.
df['DAY'] = df.DATE.dt.dayofweek
```

```python
print(df.isnull())
```

```python
df
```

```python
# Calculate sum of each day of the week and compare
daysgraph = df.groupby('DAY').sum()
daysgraph.plot(kind='bar')
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
# Weekday = [1,2,3,4,5] and Weekend = [0,6]
daysgraph = daysgraph.reset_index()
daysgraph
```

```python
# We want to view average traffic on weekday vs average traffic
# on weekend. This time, we need to groupby using mean instead of sum.
daysgraph['WEEKEND'] = daysgraph.DAY.map({0:False,1:False,2:False,3:False,4:False,5:True,6:True})
wkndgraph = daysgraph.groupby('WEEKEND').mean()
wkndgraph[['ENTRIES', 'EXITS']].plot(kind='bar')
```

```python
# Entry/Exit traffic is similar on Weekdays and Weekends.
```
